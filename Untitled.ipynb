```python
from matplotlib import pyplot as plt
import random
```

```python
x = [1, 2, 3, 4, 5]
y = [random.random() for _ in range(5)]
y2 = [random.random() for _ in range(5)]
```

```python
print(x)
print(y)
```

```python
plt.plot(x, y, label='y1')
plt.plot(x, y2, label='y2')
plt.title('random test')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
img=np.zeros((2160,4096,3))
h, w, _ = img.shape
for i in range(h):
    for j in range(w):
        img[i,j]=np.array([random.random(),random.random(),random.random()])
        
plt.imshow(img)
plt.imsave('4k.png', img)
```
