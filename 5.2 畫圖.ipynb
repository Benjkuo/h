```python
import cv2
from matplotlib import pyplot as plt
u=cv2.imread("trump.jpg")

plt.imshow(u)
```

```python
u.shape
plt.imshow(u[1:900,2:1050])
```

```python
import numpy as np
f = np.array(
[[9,-1,3],
[-9,1,3],
[-200,200,0]])

out = cv2.filter2D(u, -1, f)
plt.imshow(out)
out = np.hstack((u, out))
out=cv2.cvtColor(out,cv2.COLOR_RGB2BGR)
cv2.imwrite("goodpineapple.png", out)
```

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
u=cv2.imread("Spike.jpg")
print(u.shape)
u=cv2.cvtColor(u,cv2.COLOR_BGR2RGB)
f = np.array(
[[-1,-1,-1],
[-1,9,-1],
[-1,-1,-1]])

out = cv2.filter2D(u, -1, f)
out = np.hstack((u, out))
plt.imshow(out)
out=cv2.cvtColor(out,cv2.COLOR_RGB2BGR)
cv2.imwrite("pineapple.png", out)
```
